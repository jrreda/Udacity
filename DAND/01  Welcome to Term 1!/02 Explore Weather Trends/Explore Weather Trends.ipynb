```python
# import important packages
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# read global data
globe = pd.read_csv('global_data.csv')
globe.head()
```

```python
# read Cairo data
city = pd.read_csv('city_data.csv')
city.head()
```

# Before Moving Average

```python
# Change seaborn style
sns.set_style("darkgrid")
```

```python
# Plot global Average temprature
_ = sns.lineplot(x='year', y='avg_temp', data=globe)
```

```python
# Plot Cairo Average temprature
_ = sns.lineplot(x='year', y='avg_temp', data=city)
```

```python
## Putting the 2 plotes together and fix the y axis
plt.figure(figsize=(16, 6))

# global Average temprature
ax1 = plt.subplot(121)
sns.lineplot(x='year', y='avg_temp', data=globe, ax=ax1)
ax1.set_ylim([6, 25])
ax1.set_title('Global temprature')

# Cairo Average temprature
ax2 = plt.subplot(122)
sns.lineplot(x='year', y='avg_temp', data=city, ax=ax2)
ax2.set_ylim([6, 25])
ax2.set_title('Cairo temprature')

plt.show()
```

# After Moving Average

```python
# Create moving average temprature column in global dataframe
globe['MA_temp'] = globe.avg_temp.rolling(window=10).mean()
```

```python
# Create moving average temprature column in Cairo dataframe
city['MA_temp'] = city.avg_temp.rolling(window=10).mean()
```

```python
# Set figure size
plt.figure(figsize=(14, 5))

# global Moving Average temprature
sns.lineplot(x='year', y='MA_temp', data=globe)

# Cairo Moving Average temprature
sns.lineplot(x='year', y='MA_temp', data=city)

# Set plot title
plt.title('Cairo VS Global temprature')

# put legend on the plot
plt.legend(['Global', 'Cairo'], loc='center')

plt.show()
```
